```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# مسیر فایل‌های CSV
file1_path = 'CarPrice_Assignment.csv'

# خواندن فایل CSV
df = pd.read_csv(file1_path, encoding='ISO-8859-1')

# نمایش نمونه‌ای از داده‌ها برای بررسی
print(df.head())

# حذف ستون‌های غیر ضروری
df = df.drop(columns=['CarName'])

# تبدیل داده‌های دسته‌بندی به عددی
df = pd.get_dummies(df, drop_first=True)

# نمایش نمونه‌ای از داده‌ها برای بررسی
print(df.head())
```

```python
# هیستوگرام نوع سوخت موتور
sns.histplot(df['fueltype_gas'])
plt.xlabel('نوع سوخت')
plt.ylabel('تعداد')
plt.title('توزیع نوع سوخت موتور')
plt.show()
```

15 درصد داده را برای test و بقیه را برای train و validation استفاده کنید.
شبکه عصبی مناسبی برای این پیش‌بینی طراحی و بهینه‌سازی کرده و loss و معیارهای rmse و mae را برای 100 epoch نمایش دهید.

```python
# جداسازی ویژگی‌ها و هدف
X = df.drop(columns=['price'])
y = df['price']

# نرمال‌سازی داده‌ها
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# تقسیم داده به تست و ترین/ولیدیشن
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.15, random_state=42)

# طراحی و آموزش شبکه عصبی
def build_model(num_neurons):
    model = Sequential()
    model.add(Dense(num_neurons, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(1))  # لایه خروجی
    model.compile(loss='mean_squared_error', optimizer=Adam(), metrics=['mae', 'mse'])
    return model

# تعداد نورون‌های مختلف برای بررسی
neuron_counts = [50, 100, 200, 300, 500]

for neurons in neuron_counts:
    model = build_model(neurons)
    history = model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=32, verbose=1)
    # نمایش loss و معیارهای rmse و mae برای 100 epoch
    plt.plot(history.history['mae'], label='Train MAE')
    plt.plot(history.history['val_mae'], label='Validation MAE')
    plt.xlabel('Epoch')
    plt.ylabel('MAE')
    plt.title(f'Model with {neurons} Neurons')
    plt.legend()
    plt.show()
```

شبکه‌ای با یک لایه و نورون‌های بین 50 تا 500 برای حل این مسئله طراحی کنید. مساله را برای 5 تعداد نورون مختلف حل کنید و مینیمم mae بدست آمده را برای هر تعداد نورون در یک گراف رسم کنید. اثر افزایش تعداد نورون را تحلیل کنید.

```python
# طراحی و آموزش شبکه عصبی
def build_model(num_neurons):
    model = Sequential()
    model.add(Dense(num_neurons, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(1))  # لایه خروجی
    model.compile(loss='mean_squared_error', optimizer=Adam(), metrics=['mae', 'mse'])
    return model

# تعداد نورون‌های مختلف برای بررسی
neuron_counts = [50, 100, 200, 300, 500]
min_mae = []

for neurons in neuron_counts:
    model = build_model(neurons)
    history = model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=32, verbose=1)
    min_mae.append(min(history.history['val_mae']))

# رسم نمودار مینیمم MAE برای هر تعداد نورون
plt.plot(neuron_counts, min_mae, marker='o')
plt.xlabel('تعداد نورون')
plt.ylabel('مینیمم MAE')
plt.title('اثر افزایش تعداد نورون بر مینیمم MAE')
plt.grid(True)
plt.show()

# تحلیل اثر افزایش تعداد نورون
for i, neurons in enumerate(neuron_counts):
    print(f'تعداد نورون: {neurons}, مینیمم MAE: {min_mae[i]}')
```

شبکه‌ای با 100 نورون در هر لایه و تعداد لایه‌های 1، 2، 5 و 7 بسازید. مانند مرحله قبل، مینیمم mae بدست آمده را برای هر تعداد لایه رسم کنید. اثر افزایش تعداد لایه را تحلیل کنید.

```python
# طراحی و آموزش شبکه عصبی
def build_model(num_layers):
    model = Sequential()
    model.add(Dense(100, input_dim=X_train.shape[1], activation='relu'))
    for _ in range(num_layers - 1):
        model.add(Dense(100, activation='relu'))
    model.add(Dense(1))  # لایه خروجی
    model.compile(loss='mean_squared_error', optimizer=Adam(), metrics=['mae', 'mse'])
    return model

# تعداد لایه‌های مختلف برای بررسی
layer_counts = [1, 2, 5, 7]
min_mae = []

for layers in layer_counts:
    model = build_model(layers)
    history = model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=32, verbose=1)
    min_mae.append(min(history.history['val_mae']))

# رسم نمودار مینیمم MAE برای هر تعداد لایه
plt.plot(layer_counts, min_mae, marker='o')
plt.xlabel('تعداد لایه‌ها')
plt.ylabel('مینیمم MAE')
plt.title('اثر افزایش تعداد لایه بر مینیمم MAE')
plt.grid(True)
plt.show()

# تحلیل اثر افزایش تعداد لایه
for i, layers in enumerate(layer_counts):
    print(f'تعداد لایه‌ها: {layers}, مینیمم MAE: {min_mae[i]}')
```

برای تحلیل اثر افزایش تعداد لایه‌ها بر دقت مدل، می‌توانیم به نتایج مینیمم MAE (Mean Absolute Error) برای هر تعداد لایه نگاه کنیم و نمودار مربوطه را بررسی کنیم. به این ترتیب، متوجه می‌شویم که چگونه افزایش تعداد لایه‌ها بر دقت مدل تأثیر می‌گذارد. تحلیل اثر افزایش تعداد لایه‌ها را می‌توان به شکل زیر انجام داد:

تعداد لایه‌های کم (1 یا 2 لایه):

در این حالت، مدل ساده‌تر است و احتمالاً نمی‌تواند پیچیدگی‌های موجود در داده‌ها را به خوبی یاد بگیرد. این ممکن است منجر به خطای بیشتری شود.
مینیمم MAE ممکن است بالاتر باشد زیرا مدل قادر به فراگیری تمام الگوها و ویژگی‌های موجود در داده‌ها نیست.
افزایش تعداد لایه‌ها (5 لایه):

با افزایش تعداد لایه‌ها، مدل توانایی یادگیری ویژگی‌های پیچیده‌تر را دارد. این می‌تواند به کاهش خطا و بهبود دقت مدل کمک کند.
مینیمم MAE در این حالت ممکن است کمتر شود، زیرا مدل قادر است الگوها و روابط پیچیده‌تری را در داده‌ها یاد بگیرد.
افزایش بیشتر تعداد لایه‌ها (7 لایه):

هرچند افزایش تعداد لایه‌ها می‌تواند به بهبود دقت مدل کمک کند، اما بعد از یک نقطه خاص، افزایش بیشتر لایه‌ها ممکن است به دلیل مشکلاتی مانند overfitting (بیش‌برازش) منجر به افزایش خطا شود.
مینیمم MAE ممکن است افزایش یابد زیرا مدل به جای یادگیری الگوهای عمومی، شروع به یادگیری جزئیات و نویزهای موجود در داده‌های آموزشی می‌کند.
توجه به تعادل:

برای به دست آوردن بهترین نتیجه، باید تعادلی بین تعداد لایه‌ها و پیچیدگی مدل برقرار کرد. استفاده از تعداد لایه‌های مناسب و بهینه‌سازی های دیگر مانند regularization می‌تواند به دستیابی به بهترین دقت کمک کند.
نمودار مینیمم MAE در مقابل تعداد لایه‌ها به ما کمک می‌کند تا این تعادل را پیدا کنیم.
برای یک تحلیل دقیق‌تر، باید نمودار مینیمم MAE در مقابل تعداد لایه‌ها را رسم کنیم و نقاط بهینه را شناسایی کنیم. نمودار زیر می‌تواند به عنوان یک نمونه از تحلیل باشد:

```python
import matplotlib.pyplot as plt

# نتایج مینیمم MAE برای هر تعداد لایه
layers = [1, 2, 5, 7]
min_mae = [2000, 1800, 1500, 1600]

plt.plot(layers, min_mae, marker='o')
plt.xlabel('تعداد لایه‌ها')
plt.ylabel('مینیمم MAE')
plt.title('اثر افزایش تعداد لایه‌ها بر مینیمم MAE')
plt.show()
```

**شبکه wide and deep مناسبی برای حل این مسئله پیاده‌سازی کرده و نتایج را با شبکه deep بهینه شده مقایسه کنید.**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# ساخت مدل Deep بهینه‌شده
def build_deep_model():
    model = Sequential()
    model.add(Dense(100, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(100, activation='relu'))
    model.add(Dense(100, activation='relu'))
    model.add(Dense(100, activation='relu'))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer=Adam(), metrics=['mae', 'mse'])
    return model

deep_model = build_deep_model()
history_deep = deep_model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=32, verbose=1)

# نمایش loss و معیارهای mae برای 100 epoch
plt.plot(history_deep.history['mae'], label='Train MAE')
plt.plot(history_deep.history['val_mae'], label='Validation MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.title('Deep Model')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.layers import Input, concatenate
from tensorflow.keras.models import Model

# ساخت مدل Wide and Deep
def build_wide_and_deep_model():
    input_layer = Input(shape=(X_train.shape[1],))

    # بخش Wide
    wide_layer = Dense(100, activation='relu')(input_layer)

    # بخش Deep
    deep_layer = Dense(100, activation='relu')(input_layer)
    deep_layer = Dense(100, activation='relu')(deep_layer)
    deep_layer = Dense(100, activation='relu')(deep_layer)
    deep_layer = Dense(100, activation='relu')(deep_layer)

    # ادغام بخش‌های Wide و Deep
    combined = concatenate([wide_layer, deep_layer])

    output_layer = Dense(1)(combined)

    model = Model(inputs=input_layer, outputs=output_layer)
    model.compile(loss='mean_squared_error', optimizer=Adam(), metrics=['mae', 'mse'])
    return model

wide_and_deep_model = build_wide_and_deep_model()
history_wide_and_deep = wide_and_deep_model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=32, verbose=1)

# نمایش loss و معیارهای mae برای 100 epoch
plt.plot(history_wide_and_deep.history['mae'], label='Train MAE')
plt.plot(history_wide_and_deep.history['val_mae'], label='Validation MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.title('Wide and Deep Model')
plt.legend()
plt.show()
```

**مقایسه نتایج**:  
برای مقایسه نتایج دو مدل، می‌توان از معیارهای MAE و MSE بر روی داده‌های تست استفاده کرد:

```python
# ارزیابی مدل Deep بهینه‌شده
deep_model.evaluate(X_test, y_test)

# ارزیابی مدل Wide and Deep
wide_and_deep_model.evaluate(X_test, y_test)
```

**تحلیل نتایج**  
با مقایسه مقادیر MAE و MSE بدست آمده از دو مدل، می‌توان تحلیل کرد که کدام مدل عملکرد بهتری دارد. عموماً شبکه‌های Wide and Deep به دلیل ترکیب ویژگی‌های مختلف در شبکه‌های Wide و Deep می‌توانند عملکرد بهتری داشته باشند.
